```python
#Converting Datatypes
#Converting Orderdate to datetime
df['Orderdate'] = pd.to_datetime(
    df['Orderdate'],
    format='%m/%d/%Y',
    errors='coerce'
)

#Converting Duedate to datetime
df['Duedate'] = pd.to_datetime(
    df['Duedate'],
    format='%d-%m-%Y',
    errors='coerce'
)

#Converting Shipdate to datetime
df['Shipdate'] = pd.to_datetime(
    df['Shipdate'],
    format='%d-%m-%Y',
    errors='coerce'
)
```
